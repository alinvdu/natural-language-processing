```python
from transformers import pipeline, set_seed

import textwrap
import numpy as np
import matplotlib.pyplot as plt

from pprint import pprint
```

```python
!cat '../../data/steven_wallace.txt'
```

```python
lines = [line.rstrip() for line in open('../../data/steven_wallace.txt')]
lines = [line for line in lines if len(line) > 0]
```

```python
gen = pipeline("text-generation")
```

```python
set_seed(10)
```

```python
lines[0]
```

```python
gen(lines[0])
```

```python
pprint(_)
```

```python
pprint(gen(lines[0], max_length=20))
```

```python
pprint(gen(lines[0], num_return_sequences=3, max_length=20))
```

```python
def wrap(x):
    return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
out = gen(lines[0], max_length=30)
print(wrap(out[0]['generated_text']))
```

```python
prev = 'Pour the unhappiness out of your neighbor\'s house and bring the heat' + \
    'up to them.  This should do the trick.  Use a few spices to'
```

```python
out = gen(prev + '\n' + lines[2], max_length=60)
print(wrap(out[0]['generated_text']))
```
