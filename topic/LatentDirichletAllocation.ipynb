```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import textwrap

from nltk.corpus import stopwords

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
nltk.download("stopwords")
```

```python
stops = set(stopwords.words('english'))
```

```python
stops = stops.union({
    'said', 'would', 'told', 'also', 'one', 'two',
    'mr', 'new', 'year',
})
```

```python
df = pd.read_csv("../data/bbc_text_cls.csv")
```

```python
df.head()
```

```python
vectorizer = CountVectorizer(stop_words=list(stops))
```

```python
X = vectorizer.fit_transform(df['text'])
```

```python
lda = LatentDirichletAllocation(
    n_components=10,
    random_state=12345
)
```

```python
lda.fit(X)
```

```python
def plot_top_words(model, features_names, n_top_words=10):
    fig, axes = plt.subplots(2, 5, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[: -n_top_words - 1 : -1]
        top_features = [features_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]
        
        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f"Topic {topic_idx + 1}", fontdict={"fontsize": 30})
        ax.invert_yaxis()
        ax.tick_params(axis="both", which="major", labelsize=20)
        for i in "top right left".split():
            ax.spines[i].set_visible(False)
        fig.suptitle('LDA', fontsize=40)
```

```python
features_names = vectorizer.get_feature_names_out()
plot_top_words(lda, features_names)
```

```python
Z = lda.transform(X)
```

```python
np.random.seed(0)
i = np.random.choice(len(df))
z = Z[i]
topics = np.arange(10) + 1

fig, ax = plt.subplots()
ax.barh(topics, z)
ax.set_yticks(topics)
ax.set_title('True label: %s' % df.iloc[i]['labels'])
```

```python
def wrap(x):
    return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)

print(wrap(df.iloc[i]['text']))
```
